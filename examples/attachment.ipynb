# Firtree Attachment Example


```python
from turbodesigner.attachments.firetree import FirtreeAttachment
import numpy as np

attatchment = FirtreeAttachment(
    gamma=np.radians(40), 
    beta=np.radians(40), 
    ll=1.629, 
    lu=3.62, 
    Ri=0.55, 
    Ro=0.55, 
    R_dove=1, 
    max_length=13, 
    num_stages=2, 
    disk_radius=50
)

attatchment.visualize()
```
