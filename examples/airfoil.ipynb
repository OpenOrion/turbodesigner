# Double Circular Arc Airfoil

```python
import numpy as np
from turbodesigner.airfoils.DCA import DCAAirfoil

c = 1
theta = np.radians(40)
tb = c * 0.1
r0 = tb * 0.1
xi = np.radians(20)
airfoil = DCAAirfoil(c=c, theta=theta, r0=r0, tb=tb, xi=xi)
actual_coords = airfoil.visualize(num_arc_points=50,num_circle_points=50)
```
